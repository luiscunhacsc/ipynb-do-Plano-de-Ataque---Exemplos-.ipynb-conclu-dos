```python
# ðŸ“¦ Importar as bibliotecas necessÃ¡rias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# ðŸŽ¯ 1. Criar dados fictÃ­cios (Ã¡rea em mÂ² e preÃ§o em milhares de euros)
X = np.array([50, 60, 70, 80, 100, 120, 150]).reshape(-1, 1)  # Ãrea (variÃ¡vel independente).
y = np.array([100, 115, 130, 150, 180, 210, 260])  # PreÃ§o (variÃ¡vel dependente)

# ðŸ“ˆ 2. Criar e treinar o modelo de RegressÃ£o Linear
modelo = LinearRegression()
modelo.fit(X, y)

# ðŸ” 3. Fazer previsÃµes
area_nova = np.array([[90]])  # Nova casa com 90 mÂ²
preco_previsto = modelo.predict(area_nova)

print(f"ðŸ“¢ PrevisÃ£o: Uma casa com 90 mÂ² deverÃ¡ custar cerca de {preco_previsto[0]:.2f} mil euros.")

# ðŸŽ¨ 4. Visualizar os dados e a linha de regressÃ£o
plt.scatter(X, y, color="blue", label="Casas reais")
plt.plot(X, modelo.predict(X), color="red", label="Linha de regressÃ£o")
plt.scatter(area_nova, preco_previsto, color="green", s=100, label="PrevisÃ£o 90 mÂ²")
plt.xlabel("Ãrea (mÂ²)")
plt.ylabel("PreÃ§o (mil euros)")
plt.title("RegressÃ£o Linear - PreÃ§o vs Ãrea")
plt.legend()
plt.grid(True)
plt.show()
```
