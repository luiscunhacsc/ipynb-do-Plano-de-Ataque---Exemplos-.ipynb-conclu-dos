```python
# 📦 Importar as bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 🎯 1. Criar dados fictícios (área em m² e preço em milhares de euros)
X = np.array([50, 60, 70, 80, 100, 120, 150]).reshape(-1, 1)  # Área (variável independente).
y = np.array([100, 115, 130, 150, 180, 210, 260])  # Preço (variável dependente)

# 📈 2. Criar e treinar o modelo de Regressão Linear
modelo = LinearRegression()
modelo.fit(X, y)

# 🔍 3. Fazer previsões
area_nova = np.array([[90]])  # Nova casa com 90 m²
preco_previsto = modelo.predict(area_nova)

print(f"📢 Previsão: Uma casa com 90 m² deverá custar cerca de {preco_previsto[0]:.2f} mil euros.")

# 🎨 4. Visualizar os dados e a linha de regressão
plt.scatter(X, y, color="blue", label="Casas reais")
plt.plot(X, modelo.predict(X), color="red", label="Linha de regressão")
plt.scatter(area_nova, preco_previsto, color="green", s=100, label="Previsão 90 m²")
plt.xlabel("Área (m²)")
plt.ylabel("Preço (mil euros)")
plt.title("Regressão Linear - Preço vs Área")
plt.legend()
plt.grid(True)
plt.show()
```
